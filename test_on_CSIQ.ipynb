```python
import cv2

def get_images(dst_type , dst_level):
    imgs = []
    type = glob.glob('CSIQ/dst_imgs/'+ dst_type +'/*.png')
    for im in type:
        if int(im.split('.')[-2]) == dst_level:
            imgs.append(im)
    
    return imgs


def get_dmos(dst_type , dst_level):
    dmos = []
    f = open("CSIQ/scores.txt", "r")
    lines = f.readlines()
    for l in lines:
        if int(l.split(' ')[-2].split('.')[-2]) == dst_level:
            if l.split(' ')[-2].split('.')[-3] == dst_type:
                dmos.append(float(l.split(' ')[-1][:5]))

    return dmos
        
    

```

```python
from scipy import stats


def get_srocc_plcc(rest , deg , dmos):
    preds = []
    gt = []
    for i in range(len(deg)):
        pred = iqa.predict(restored_addr = rest[i] , degraded_addr= deg[i])
        preds.append(pred)

    # print("rest , deg = " , len(rest) , len(deg))
    # print("pred , dmos = " , len(preds) , len(dmos))
    plcc = stats.pearsonr(preds, dmos)
    srcc = stats.spearmanr(preds, dmos)
    return srcc , plcc
        
        
```
